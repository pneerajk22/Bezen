```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('2022_02_08-02_30_31_AM.csv')
```

# 1:-Products Without and With  Prices
### without:- That is Neither Price_string or Price_string_unf is having a Value
### with:- That is eithet of them having a value


```python
with_prices = data.loc[data['price_string'].notnull() | data['price_string_unf'].notnull()]
without_prices = data.loc[data['price_string'].isna() & data['price_string_unf'].isna()]
print(f'Total rows with prices:- ',len(with_prices),'\t Total rows without prices:- ',len(without_prices))
```

# Graph demonstarting No.Of rows with prices and without prices

```python
wp = len(with_prices)
wnp = len(without_prices)
dt = {'tasks':[wp,wnp]}
wpp_df = pd.DataFrame(dt,columns=['tasks'],index = ['with_prices','without_prices'])
wpp_df.plot.pie(y='tasks',figsize=(5, 5),autopct='%1.1f%%', startangle=90)
plt.show()
```

# 2:-Count of products

### with prices and without prices in each product type, category and level_1

## Count by each product_type, with and without prices.

### Here we can use value_counts() function to count the unique value repeating in whole table

```python
with_price_pdtp = with_prices['product_type'].value_counts()
without_prices_pdtp = without_prices['product_type'].value_counts()
df = pd.merge(with_price_pdtp, without_prices_pdtp,left_index = True, right_index = True)
df = df.reset_index()
df.columns = ['product_type','With_Prices','Without_Prices']
df

```

## Count by each Category, with and without Prices

```python
with_ctgr = with_prices['category'].value_counts()
without_ctgr = without_prices['category'].value_counts()
df1 = pd.merge(with_ctgr , without_ctgr,left_index = True, right_index = True)
df1 = df1.reset_index()
df1.columns = ['category','With_Prices','Without_Prices']
df1
```

## Count by each Level_1, with and without Prices

```python
with_lvl = with_prices['level_1'].value_counts()
without_lvl = without_prices['level_1'].value_counts()
df2 = pd.merge(with_lvl , without_lvl,left_index = True, right_index = True)
df2 = df2.reset_index()
df2.columns = ['level_1','With_Prices','Without_Prices']
df2

```

## Creating a price 'values' copy from with_prices dataframe 

```python
clean_values = with_prices['price_string'].replace({'\$': '', ',': ''}, regex=True).astype(float)
```

```python
clean_values
```

## Lets create a copy dataframe from main with_prices df to eleminate any errors accouring to main data which effects only copied df

```python
new_df = with_prices[['uuid','product_type','level_1','category']]
new_df
```

## lets add the currency and value columns to the new_df

```python
new_df.insert(1, 'currency', '$')
```

```python
new_df.insert(2, 'value', clean_values)
new_df
```

# 4:- list out categories with avg price of products

```python
new_df1 = new_df[['category','value']]
new_df1
```

```python
new_df2 = new_df1.groupby(['category']).agg({'value': 'mean'})
new_df2
```

```python
new_df2.rename({'value': 'product_price_avg'},axis=1)
```
